```python
import pandas as pd
import gdown
from google.colab import files

#Sentiment just for fun, going to use VADER
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')

pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', False)
pd.set_option('display.max_rows', 200)
```

```python
archive_text = pd.read_csv("web-pages_full.csv")
```

```python
archive_text['crawl_date']= pd.to_datetime(archive_text['crawl_date'],format='%Y%m%d')
archive_text = archive_text[archive_text['language'] =='en']
```

```python
archive_text.groupby(["domain"]).count().sort_values(by="crawl_date", ascending=False)[0:50]
```

```python
#Municiple domains into a specific dataframe

municiple_domains = [

  "www.niagararegion.ca",
  "www.notl.com",
  "www.stcatharines.ca",
  "niagarafalls.ca",
  "lincoln.ca",
  "www.forterie.ca",
  "www.grimsby.ca",
  "portcolborne.ca",
  "www.thorold.ca",
  "www.welland.ca",
  "westlincoln.ca",
  "www.wainfleet.ca",
  "www.pelham.ca"
]

muni_text = archive_text[archive_text.domain.isin(municiple_domains)]
```

```python
#Add page length
for index, row in muni_text.iterrows():
    muni_text.loc[index,'length'] = int(len(row['content']))
```

```python
#VADER Scoring

muni_text.loc[:,"vscore_pos"] = 0.0
muni_text.loc[:,"vscore_neg"] = 0.0
muni_text.loc[:,"vscore_neu"] = 0.0
muni_text.loc[:,"vscore_compound"] = 0.0


sid = SentimentIntensityAnalyzer()
for index, row in muni_text.iterrows():
    ss = sid.polarity_scores(row["content"])
    muni_text.loc[index,'vscore_pos'] = float(ss["pos"])
    muni_text.loc[index,'vscore_neg'] = float(ss["neg"])
    muni_text.loc[index,'vscore_neu'] = float(ss["neu"])
    muni_text.loc[index,'vscore_compound'] = float(ss["compound"])
```

```python
#write of final CSV and download

muni_text.to_csv("municiple_text.csv")
files.download("municiple_text.csv")
```
